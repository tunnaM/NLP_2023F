```python
import pandas as pd
from snownlp import SnowNLP
```

```python
data = pd.read_csv('瑞金医院其他数据.csv',header=None
                   #,encoding='gbk'
                  )
data
```

```python
data[11]=''
```

```python
for i in range(0,len(data)):
  # 情感得分
  try:
    emo = SnowNLP(data[10][i].strip()).sentiments
    data[11][i] = emo
  except: 
    print('异常')
print('write down.')
```
