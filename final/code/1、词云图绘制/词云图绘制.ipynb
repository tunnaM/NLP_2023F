```python
import re
import jieba
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
def sw(filename):
    with open(filename, encoding="utf-8") as file:
        x = file.readlines()
    y = [word.strip() for word in x]
    return y
```

```python
def get_varname(var, locals=locals()):
    for k, v in locals.items():
        if v is var:
            return k     
```

```python
def draw_wc(datalist):
    # 导入中文停用词
    name_list = ['stopword/baidu_stopwords.txt', 'stopword/cn_stopwords.txt', 
            'stopword/hit_stopwords.txt','stopword/scu_stopwords.txt','stopword/select.txt']
    stop_word = []
    for x in name_list:
        stop_word.extend(sw(x))
    stop_word = list(set(stop_word))
    # 中英文分词
    sentence = datalist['政策名称'].tolist()
    sentence.extend(datalist['政策内容'].tolist())
    txt_c = ""
    counts = {}
    background_image= np.array(Image.open(r'E:\0、大四上\2、文本挖掘 项目\项目\词云图绘制\医院.png'))
    # image_colors = ImageColorGenerator(msk)
    color_list=["#5B9BD5","#4472C4"]
    colormap=colors.ListedColormap(color_list)
    for i in range(0,len(sentence)):
        line = ''.join(re.findall('[\u4e00-\u9fa5]',str(sentence[i])))
        if line:
            c = jieba.lcut(line)
            result = [word for word in c if word not in stop_word]
            c = [word for word in result if len(word)>1]
            txt_c += " ".join(c)
            # 词频统计
            for c_i in c:
                if len(c_i) == 1:
                    continue
                else:
                    counts[c_i] = counts.get(c_i,0) + 1
    # 词频统计处理
    items = list(counts.items())
    items.sort(key=lambda x:x[1], reverse=True)
    # 绘制中文词云图
    font = r'C:\Windows\Fonts\simkai.ttf'
    wc = WordCloud(font_path=font,
                   background_color='white',
                   width=1000,
                   height=800,
                  mask=background_image, # 指定词云的形状
                  colormap=colormap
                  ).generate(txt_c)
    wc.to_file(get_varname(datalist) + '.png')
    # plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
    # plt.axis("off")
    # plt.show()
    return items
```
